## 결측치 처리

### Drop
   #### 경우에 따라 결측치 처리 방법이 달라진다.
   #### 샘플수가 많다면 missing values 를 포함하는 행을 모두 삭제하는 것이 가능하다

```python
# 결측치가 하나라도 있으면 버리는 코드 예제
df.dropna()

# 모든 값이 Null일 경우만 버리는 코드 예제
df.dropna(how='all')

# 결측치가 하나 이상 있는 Case만 선택하는 코드 예제
df[df.isnull().any(axis=1)]
```

### Imputation
#### 만약 샘플수가 충분하지 않을 경우, Pandas의 fillna() 명령어로 Null 값을 채우는 것이 가능하다.
#### 연속형인 경우 Mean이나 Median을 이용하고 명목형인 경우 Mode(최빈치)나 예측 모형을 통해 Null 값을 대체할 수 있다.

```python
df_by_screen = df.groupby(['datetime', 'screen'])['sessionid'].nunique().unstack()
df_by_screen.head()
```

```python
df_by_screen.isnull().sum()
```

```python
df_by_screen.info()
```

#### 비즈니스 맥락 및 목적(퍼널분석)을 고려했을 때, 이 경우는 결측치를 0으로 처리하는 것이 적당하다고 판단됨.(구매를 안했을 수도 있다)

```python
df_by_screen.fillna(0)
```

```python
df_by_screen.dropna(how='any').head()
```

```python
df_by_screen.dropna(how='all').head()
```

```python
df_by_screen[df_by_screen.isnull().any(axis = 1)]
```

```python
df_by_screen.fillna(df_by_screen.median()).head()
```

## 2.4 결측치가 카테고리 변수인 경우
### Option 1. 최빈치(mode)쉽고 빠르게 이용가능하나, 낮은 적절성<br>Option 2. 예측모형으로 예측치<br>적절성은 비교적 높으나 별도의 모형과 계산량 필요

```python
df['documentposition'][0]
```

```python
df_ms = df.copy()
```

```python
df_ms.iloc[2]['documentposition'] = np.nan
df_ms.iloc[5]['documentposition'] = np.nan
df_ms.iloc[7]['documentposition'] = np.nan
df_ms.iloc[10]['documentposition'] = np.nan
df_ms.iloc[11]['documentposition'] = np.nan
df_ms.iloc[15]['documentposition'] = np.nan
```

```python
df_ms.head()
```

```python
df_ms.documentposition.value_counts()
```

```python
# fillna with most frequent value (MODE)
freq_value = df_ms.documentposition.value_counts().index[0]
freq_value
```

```python
df_ms.fillna(freq_value)
```
